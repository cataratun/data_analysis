```python
import pandas as pd

insta = pd.read_csv('/content/drive/MyDrive/dataAnalysis/Instagram.csv')

#insta.info()
#imprime los tipos de datos originales

insta["reg date"] = pd.to_datetime(insta["reg date"])
  #cambia el tipo de dato de "reg date" a fecha

insta["last login"] = pd.to_datetime(insta["last login"])
  #cambia el tipo de dato de "last login" a fecha

insta["gender"] = insta["gender"].astype("category")
  #cambia el tipo de dato de "gender" a categoría

insta["Country"] = insta["Country"].astype("category")
  #cambia el tipo de dato de "Country" a categoría

insta["Influencer"] = insta["Influencer"].astype("bool")
  #cambia el tipo de dato de "Influencer" a booleano

insta.info()
#imprime los cambios para hacerlos visibles
```

```python
import pandas as pd

insta = pd.read_csv('/content/drive/MyDrive/dataAnalysis/Instagram.csv', parse_dates = ["reg date", "last login"])
#cambia los tipos de datos de "reg date" y "last login" a fecha

#insta.info()

insta["gender"] = insta["gender"].astype("category")
insta["Country"] = insta["Country"].astype("category")
  #cambia "gender" y "country" a tipo de dato categoría

insta["Influencer"] = insta["Influencer"].astype("bool")
  #cambia "Influencer" a tipo de dato booleano

insta.info()
#imprime la información de los tipos de datos de cada columna con los cambios
```

```python
insta[insta["gender"] == "Male"]
insta[insta["Country"] == "Australia"]
#imprime las filas en las que "gender" coincida con "Male" o "Country" coincida con "Australia"
```

```python
mask = insta["gender"] == "Female"
insta[mask]
#devuelve las filas en las que "gender" coincida con "Female"
```
