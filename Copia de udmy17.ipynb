```python
df = pd.DataFrame(columns = fortune.columns)
  #crea una copia de df vacía
df
```

```python
dfs = []
# lista vacía para almacenar los resultados

for sector, values in sectors:
    company_with_highest_profit = values.nlargest(1, "Profits")
      #selecciona la fila con mayor ganancia o mayor valor en "Profits"
    dfs.append(company_with_highest_profit)
      #agrega cada fila a la lista

df = pd.concat(dfs, ignore_index=True)
#combina todo en un solo dataframe

df
#imprime el dataframe combinado
```
