```python
reg_course = students_rec1.groupby('course_name')
#agrupa por "couse_name"
reg_course.size().sort_values()
#imprime el tamaño de cada valor contenido en "reg_course" agrupado en "course name" y los ordena
#imprime el tipo de dato
```
