```python
import pandas as pd
```

```python
numbers = [1, 2, 3, 4, 5, 6]
#Crear un DataFrame a partir de una lista
```

```python
pd.DataFrame(numbers)
#Convierte la lista en DataFrame de una sola columna con sus índices
```

```python
data = [["Jhon",  23], ["Mary", 18], ["Jhon", 31]]
#Una nueva lista con listas internas
```

```python
pd.DataFrame(data)
#Transformada en DataFrame, agrega índice por defecto a las columnas
```

```python
pd.DataFrame(data, columns = ["name", "age"])
#Cambia los nombres asignados por defecto a las columnas
```

```python
pd.DataFrame(data, columns = ["name", "age"], index = ["id1", "id2", "id3"])
#Modifica los índices de las filas, este debe coincidir con el número de filas que hay
```
